```python
M=100000
thread,rij,xi,cont,prob=[np.zeros(M) for _ in range(5)]
with open("../testing.dat","r") as testfile:
    for l,line in enumerate(testfile):
        if l<M:
            split_line=line.split("\t")
            thread[l]=int(split_line[0])
            rij[l]=float(split_line[1])
            xi[l]=float(split_line[2])
            cont[l]=float(split_line[3])
            prob[l]=float(split_line[4].split("\n")[0])
```

```python
for th in range(4):
    this_xi=xi[np.where(thread==th)]
    plt.hist(this_xi,bins=10,alpha=0.5);
```

```python
bin2=np.where(thread==1)
plt.scatter(rij[bin2],xi[bin2],s=1,c=prob[bin2])
plt.yscale('log');
plt.ylim([0.0001,1])
plt.colorbar();
```
