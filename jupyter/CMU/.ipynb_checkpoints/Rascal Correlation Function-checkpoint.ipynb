Read in QPM and Linear functions taken from Rascal

```python
qpm_xi = np.genfromtxt('../../xi_functions/QPM_Mean_xi.txt',skip_header=2)
qpm_xi_r = np.genfromtxt('../../xi_functions/QPM_Mean_xi.txt',max_rows=1)
qpm_xi_mu = np.genfromtxt('../../xi_functions/QPM_Mean_xi.txt',skip_header=1,max_rows=1)
```

```python
linear_xi = np.genfromtxt('../../xi_functions/linear_xi.txt',skip_header=2)
```

```python
linear_xi_r = np.genfromtxt('../../xi_functions/linear_xi.txt',max_rows=1)
linear_xi_mu = np.genfromtxt('../../xi_functions/linear_xi.txt',skip_header=1,max_rows=1)
```

```python
qpm_mean = qpm_xi.mean(axis=1)
linear_mean = linear_xi.mean(axis=1)
```

Display mean correlation functions over $\mu$

```python
plt.plot(qpm_xi_r,qpm_mean*qpm_xi_r**2.,label='QPM Mean')
plt.plot(qpm_xi_r,linear_mean*qpm_xi_r**2.,label='Linear');FS=16
plt.ylabel(r'$r^2\xi$',fontsize=FS);plt.xlabel(r'$r$/Mpc',fontsize=FS);
plt.title(r"$\xi$ correlation functions averaged over $\mu$",fontsize=FS);plt.legend();
```

Rescale linear theory by mean(QPM/linear) over 40-60 Mpc/h

```python
r_index=np.where((qpm_xi_r<60)&(qpm_xi_r>40))
rescale=np.mean(qpm_mean[r_index]/linear_mean[r_index])
```

```python
print("Rescaling: Linear -> %.2f*Linear"%rescale)
```

Now replot

```python
mu_index=0
plt.plot(qpm_xi_r,qpm_mean*qpm_xi_r**2.,label='QPM Mean')
plt.plot(qpm_xi_r,rescale*linear_mean*qpm_xi_r**2.,label='Linear');FS=16
plt.ylabel(r'$r^2\xi$',fontsize=FS);plt.xlabel(r'$r$/Mpc',fontsize=FS);
plt.title(r"$\xi(r)$ correlation functions averaged over $\mu$",fontsize=FS);plt.legend();
```

```python
cols=['r','k','b','g']
for m,mu_index in enumerate([0,30,60,90]):
    plt.plot(qpm_xi_r,qpm_xi[:,mu_index]*qpm_xi_r**2.,label='%.1f'%(mu_index/100.),ls='-',c=cols[m])
    plt.plot(qpm_xi_r,rescale*linear_xi[:,mu_index]*qpm_xi_r**2.,ls='--',c=cols[m]);FS=16
plt.ylabel(r'$r^2\xi$',fontsize=FS);plt.xlabel(r'$r$/Mpc',fontsize=FS);
plt.title("QPM (dashed) vs Linear (full)",fontsize=FS);plt.legend();
```
