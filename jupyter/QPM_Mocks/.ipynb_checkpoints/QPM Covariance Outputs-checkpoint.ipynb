```python
def KL_div(prec,cov):
    logdetPrec = np.linalg.slogdet(prec)
    if logdetPrec[0]<0:
        raise Exception("Uninvertable precision matrix")
    logdetCov = np.linalg.slogdet(cov)
    if logdetCov[0]<0:
        raise Exception("Uninvertable covariance matrix")
    n_bins = len(prec)
    return 0.5*(np.trace(np.matmul(prec,cov))-logdetPrec[1]-logdetCov[1]-n_bins)
```

```python
def n_eff_KL(prec,cov):
    KL = KL_div(prec,cov)
    n_bins = len(prec)
    return n_bins*(n_bins+1.)/(4.*KL)
```

```python
n_eff_KLs=[]
for mock_no in range(min_mock,max_mock+1):
    n_eff_KLs.append(n_eff_KL(Psi,c_tots[mock_no-1]))
```
