```python
M=10000
rij,rik,rkl,cont=[np.zeros(M) for _ in range(4)]
with open("../../BOSS_DR12_Data/v8/testing.dat","r") as testfile:
    for l,line in enumerate(testfile):
        if l<10000:
            split_line=line.split("\t")
            rij[l]=float(split_line[0])
            rik[l]=float(split_line[1])
            rkl[l]=float(split_line[2])
            cont[l]=float(split_line[3].split("\n")[0])
```
