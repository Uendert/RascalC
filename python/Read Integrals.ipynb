```python
%pylab inline
import numpy as np
import csv
file_root='../CovMatrices/'
n=10;m=6;a=1.05
```

```python
def read_c_tot(n,m,file_root,a):
    c2file=file_root+'c2_n%d_m%d.txt' %(n,m)
    c3file=file_root+'c3_n%d_m%d.txt' %(n,m)
    c4file=file_root+'c4_n%d_m%d.txt' %(n,m)
    c2=np.diag(np.loadtxt(c2file))
    c3=np.loadtxt(c3file)
    c4=np.loadtxt(c4file)
    c_tot=c4;#a**2.*c2+a*c3+c4
    return c_tot
```

```python
c_tot=read_c_tot(n,m,file_root,1.)
```

```python
precision_tot=np.linalg.inv(c_tot)
```

```python
x_all=np.arange(0,len(c_tot))
import cmocean
X,X=np.meshgrid(x_all,x_all)
import matplotlib.pyplot as plt
cbar=cmocean.cm.balance
plt.matshow(c_tot)
maxval=min(np.abs(precision_tot.ravel()))
plt.title('Covariance')
plt.colorbar()
plt.figure();
plt.matshow(precision_tot,cmap=cbar)#,vmin=-maxval,vmax=maxval)
plt.colorbar()
plt.title('Precision');
```
