In uniform survey limit measured isotropic power scales as

$$P^a = 4\pi\int_0^\infty x^2 dx \xi(x) W(x) K^a(x)$$

for kernel $K^a$ in bin $a$ and window function $W(x)$. This estimates the power spectrum $P(k)*\tilde W(k)$. For the isotropic sky we have kernel

$$K^a(x) = \frac{3}{x\left(k_{a,\mathrm{max}}^3-k_{a,\mathrm{min}}^3\right)}\left[k^2j_1(kx)\right]_{k_{a,\mathrm{min}}}^{k_{a,\mathrm{max}}}$$

In the thin-bin limit, with bin-center $k_a$ we have the kernel
$$K^a(x) \rightarrow j_0(kx)$$

Consider truncation errors in this from using non-infinite $W(x)$, from a variety of window functions.

#### Define kernel

```python
from scipy.special import spherical_jn
def kernel(x,k_min,k_max):
    return 3./(x*(k_max**3.-k_min**3.))*(k_max**2.*spherical_jn(1,k_max*x)-k_min**2.*spherical_jn(1,k_min*x))
def thin_kernel(x,k_cen):
    return spherical_jn(0,k_cen*x)
```

#### Define window functions

```python
def windower(r,R_max,f,norm):
    """General window function operator for function f with support on [0,R]"""
    if type(r)==float:
        if r<R_max:
            return f(r/R_max)/norm
        else:
            return 0.
    else:
        output = np.zeros_like(r)
        output[r<R_max]=f(r[r<R_max]/R_max)/norm
        return output

from scipy.integrate import quad
def norm_fn(f,R_max):
    """Compute normalization: 4 pi int_0^R_max r^2 dr f(r/R) = 4 pi R^3 int_0^1 f(x)*x^2 dx """
    tmp_int = quad(lambda x: f(x)*x**2.,0,1)
    return tmp_int[0]*4.*np.pi*R_max**3.
```

```python
from scipy.special import i0
top_hat = lambda x: 1
neg_exp = lambda x: np.exp(-x)-1./np.e
pos_exp = lambda x: np.e-np.exp(x)
w_tanh = lambda x: np.tanh(1)-np.tanh(x)
w_arctan = lambda x: np.arctan(1)-np.arctan(x)
w_cos = lambda x: np.cos(x*np.pi/2)
w_arsinh = lambda x: np.arcsinh(1)-np.arcsinh(x)
recip_x = lambda x: 1./(1.+x)-1./2.
recip_x2 = lambda x: 1./(1.+x**2.)-1./2.
w_gauss = lambda x: np.exp(-x**2./2.)-1./np.sqrt(np.e)
triang = lambda x: 1.-x
welch = lambda x: 1.-x**2.
hann = lambda x: (np.sin(np.pi/2.*(1-x)))**2.
def kaiser(x,alpha_kaiser=3):
    return i0(alpha_kaiser*np.pi*np.sqrt(1.-x**2.))/i0(alpha_kaiser*np.pi)
```

#### Consider spherical fourier transforms of window functions

```python
from scipy.special import spherical_jn
def spherical_FT(k,R_max,f,norm):
    out=quad(lambda r: spherical_jn(0,k*r)*r**2.*windower(r,R_max,f,norm),0,R_max)
    return out[0]*np.pi*4.
```

```python
all_k = np.logspace(-2,-1,50)
for i,window in enumerate(windows):
    if names[i] in good_window:
        norm = norm_fn(window,R_max)
        fts = [spherical_FT(kk,R_max,window,norm) for kk in all_k]
        plt.plot(all_k,fts,label=names[i])
plt.legend();
plt.ylabel('FT(k)')
plt.xlabel('k')
plt.xscale('log');
```

#### Now consider integrated kernel

```python
def integrated_kernel(k_min,k_max,R_max,f,norm):
    from scipy.integrate import quad
    integ_func = lambda x: r2xi_spline(x)*(k_max**2.*spherical_jn(1,k_max*x)-k_min**2.*spherical_jn(1,k_min*x))/x*windower(x,R_max,f,norm)
    out = quad(integ_func,0,R_max)[0]*3./(k_max**3.-k_min**3.)
    return out
def integrated_kernel_thin(k_cen,R_max,f,norm):
    from scipy.integrate import quad
    integ_func = lambda x: r2xi_spline(x)*spherical_jn(0,k_cen*x)*windower(x,R_max,f,norm)
    out = quad(integ_func,0,R_max)[0]
    return out
```

```python
k_cen = 0.1 ## test k to use here
R_maxs = np.linspace(0,400,200)
for i,window in enumerate(windows):
    if names[i] in good_window:
        norm = norm_fn(window,R_max)
        ints = np.asarray([integrated_kernel_thin(k_cen,RR,window,norm) for RR in R_maxs])
        true_ints = integrated_kernel_thin(k_cen,10000,window,norm)
        plt.plot(R_maxs,(ints-true_ints)/true_ints,label=names[i])
plt.legend();
plt.ylabel('Fractional Error',fontsize=16);
plt.title(r'$\xi$-integral Truncation Error at $k_\mathrm{cen} = 0.1$',fontsize=16)
plt.xlabel(r'$R_\mathrm{max}$',fontsize=16);
#plt.xscale('log');
```

#### Now integrate without $\xi(r)$ factor

```python
def integrated_kernel_unmod(k_min,k_max,R_max,f,norm):
    from scipy.integrate import quad
    integ_func = lambda x: x**2.*(k_max**2.*spherical_jn(1,k_max*x)-k_min**2.*spherical_jn(1,k_min*x))/x*windower(x,R_max,f,norm)
    out = quad(integ_func,0,R_max)[0]*3./(k_max**3.-k_min**3.)
    return out
def integrated_kernel_unmod_thin(k_cen,R_max,f,norm):
    from scipy.integrate import quad
    integ_func = lambda x: x**2.*spherical_jn(0,k_cen*x)*windower(x,R_max,f,norm)
    out = quad(integ_func,0,R_max)[0]
    return out
```
