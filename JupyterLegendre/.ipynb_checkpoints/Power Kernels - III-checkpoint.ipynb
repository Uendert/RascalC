In uniform survey limit measured isotropic power scales as

$$P^a = 4\pi\int_0^\infty x^2 dx \xi(x) W(x) K^a(x)$$

for kernel $K^a$ in bin $a$ and window function $W(x)$. This estimates the power spectrum $P(k)*\tilde W(k)$. For the isotropic sky we have kernel

$$K^a(x) = \frac{3}{x\left(k_{a,\mathrm{max}}^3-k_{a,\mathrm{min}}^3\right)}\left[k^2j_1(kx)\right]_{k_{a,\mathrm{min}}}^{k_{a,\mathrm{max}}}$$

In the thin-bin limit, with bin-center $k_a$ we have the kernel
$$K^a(x) \rightarrow j_0(kx)$$

In the general case we can show that this is a difference of two Hankel transforms;

$$P^a(k) = \frac{3}{k_{a,\mathrm{max}}^3-k_{a,\mathrm{min}}^3}\sqrt{\frac{\pi}{2}}\left[\mathrm{Hankel}\left(\frac{3}{2},k^{-3/2}x^{-5/2}\left(x^2\xi(x)\right)W(x)\right)\right]_{k_{a,\mathrm{min}}}^{k_{a,\mathrm{max}}}$$

Consider truncation errors in this from using non-infinite $W(x)$.

#### First load in $\xi(r)$ model

```python
from scipy.interpolate import UnivariateSpline
xi_file = '/mnt/store1/oliverphilcox/Legendre2PCF/mean_xi.xi'
xi_model = np.loadtxt(xi_file,skiprows=2).mean(axis=1)
xi_r = np.genfromtxt(xi_file,max_rows=1)
xi_fit = UnivariateSpline(xi_r,xi_r**2.*xi_model)
r_max = max(xi_r)-10

def r2xi_spline(r):
    if type(r)==float:
        if r<r_max:
            return xi_fit(r)
        else:
            return xi_fit(r_max)*(r_max/r)**2.
    else:
        xi_out = np.zeros_like(r)
        filt = np.where(r<r_max)
        filt2 = np.where(r>=r_max)
        xi_out[filt] = xi_fit(r[filt])
        xi_out[filt2] = xi_fit(r_max)*(r_max/r[filt2])**2.
        return xi_out
```

```python
r_all = np.linspace(0,400,1000)
plt.plot(r_all,r2xi_spline(r_all))
plt.xlabel(r'$r$ [Mpc/h]',fontsize=16);plt.ylabel(r'$r^2\xi(r)$',fontsize=16);
```

#### Define kernel

```python
from scipy.special import spherical_jn
def kernel(x,k_min,k_max):
    return 3./(x*(k_max**3.-k_min**3.))*(k_max**2.*spherical_jn(1,k_max*x)-k_min**2.*spherical_jn(1,k_min*x))
def thin_kernel(x,k_cen):
    return spherical_jn(0,k_cen*x)
```

```python
k_cen = 0.1;r_all = np.linspace(0,600,1000)
plt.plot(r_all,kernel(r_all,0.95*k_cen,1.05*k_cen)/thin_kernel(0,k_cen),label=r'$\Delta \log k = 0.05$')
plt.plot(r_all,thin_kernel(r_all,k_cen)/thin_kernel(0,k_cen),label=r'Thin bin');
plt.ylim([-0.3,1.1]);plt.xlim([0,500]);plt.legend();
plt.title(r'Kernel Function for $k=%s$'%k_cen,fontsize=16);plt.ylabel(r'$K^a(r)$',fontsize=16);plt.xlabel('$r$ [Mpc/h]',fontsize=16);
```

#### Now consider integrated kernel

```python
from hankel import HankelTransform
def window(x,R_max=R_max):
    if type(x)==float:
        if x<R_max:
            return 1.
        else:
            return 0.
    else:
        output=np.zeros_like(x)
        output[x<R_max]=1.
        return output
```

```python
def integrated_kernel2(k_min,k_max,R_max):
    from scipy.integrate import quad
    integ_func = lambda x: r2xi_spline(x)*(k_max**2.*spherical_jn(1,k_max*x)-k_min**2.*spherical_jn(1,k_min*x))/x
    out = quad(integ_func,0,R_max)[0]*3./(k_max**3.-k_min**3.)
    return out
def integrated_kernel2_thin(k_cen,R_max):
    from scipy.integrate import quad
    integ_func = lambda x: r2xi_spline(x)*spherical_jn(0,k_cen*x)
    out = quad(integ_func,0,R_max)[0]
    return out
```

```python
for k in [0.01,0.03,0.1,0.3]:
    R_maxs = np.linspace(0,400,200)
    thin_int = np.asarray([integrated_kernel2_thin(k,RR) for RR in R_maxs])
    thin_int_exact = integrated_kernel2_thin(k,10000)
    plt.plot(R_maxs,(thin_int-thin_int_exact)/thin_int_exact,label=r'$k = %s$'%k)
plt.legend();
plt.ylabel('Fractional Error',fontsize=16);plt.xlabel(r'$R_\mathrm{max}$',fontsize=16);
plt.title('Thin $k$-bin integrated kernel with finite support');
```

```python
f=0.5
for k in [0.01,0.03,0.1,0.3]:
    R_maxs = np.linspace(0,400,200)
    this_int = np.asarray([integrated_kernel2((1-f)*k,(1+f)*k,RR) for RR in R_maxs])
    int_exact = integrated_kernel2((1-f)*k,(1+f)*k,10000)
    plt.plot(R_maxs,(this_int-int_exact)/int_exact,label=r'$k = %s$'%k)
plt.legend();
plt.ylabel('Fractional Error',fontsize=16);plt.xlabel(r'$R_\mathrm{max}$',fontsize=16);
plt.title('$\Delta\log(k) = %s$ $k$-bin integrated kernel with finite support'%f);
```

```python
f=0.5
c=['r','b','g','c']
for i,k in enumerate([0.01,0.03,0.1,0.3]):
    R_maxs = np.linspace(0,400,200)
    thin_int = np.asarray([integrated_kernel2_thin(k,RR) for RR in R_maxs])
    thin_int_exact = integrated_kernel2_thin(k,10000)
    this_int = np.asarray([integrated_kernel2((1-f)*k,(1+f)*k,RR) for RR in R_maxs])
    int_exact = integrated_kernel2((1-f)*k,(1+f)*k,10000)
    plt.plot(R_maxs,(thin_int-thin_int_exact)/thin_int_exact,label=r'$k = %s$'%k,c=c[i])
    plt.plot(R_maxs,(this_int-int_exact)/int_exact,ls=':',c=c[i])
plt.legend();
plt.ylabel('Fractional Error',fontsize=16);plt.xlabel(r'$R_\mathrm{max}$',fontsize=16);
plt.title('Thin (full) vs Thick (dashed) $\Delta\log{k} = %s$ $k$-bin integrated kernel with finite support'%f);
```

```python
def integrated_kernel_simple(k_min,k_max,R_max):
    from scipy.integrate import quad
    integ_func = lambda x: x**2.*(k_max**2.*spherical_jn(1,k_max*x)-k_min**2.*spherical_jn(1,k_min*x))/x*(np.exp(-x/R_max)-1./np.e)
    out = quad(integ_func,0,R_max)[0]*3./(k_max**3.-k_min**3.)
    return out
def integrated_kernel_simple_thin(k_cen,R_max):
    from scipy.integrate import quad
    integ_func = lambda x: x**2.*spherical_jn(0,k_cen*x)*(np.exp(-x/R_max)-1./np.e)
    out = quad(integ_func,0,R_max)[0]
    return out
```

```python
k_cen=0.1
R_max=100
r_all = np.linspace(0,R_max,1000)
integ_func = lambda x: x**2.*spherical_jn(0,k_cen*x)*(np.exp(-(x/R_max)**2.)-1./np.e)
plt.plot(r_all,integ_func(r_all))
```

```python
f=0.5
c=['r','b','g','c']
for i,k in enumerate([0.01,0.03,0.1,0.3]):
    R_maxs = np.linspace(0,400,200)
    thin_int = np.asarray([integrated_kernel_simple_thin(k,RR) for RR in R_maxs])
    this_int = np.asarray([integrated_kernel_simple((1-f)*k,(1+f)*k,RR) for RR in R_maxs])
    plt.plot(R_maxs,thin_int,label=r'$k = %s$'%k,c=c[i])
    plt.plot(R_maxs,this_int,ls=':',c=c[i])
plt.legend();
plt.ylabel('Kernel',fontsize=16);plt.xlabel(r'$R_\mathrm{max}$',fontsize=16);
plt.title('Thin (full) vs Thick (dashed) $\Delta\log{k} = %s$ $k$-bin integrated kernel with finite support'%f);
```

#### Think about $~1/x^2$ kernel

```python
R_max = 200
from scipy.integrate import quad
ft_window = lambda k: 4.*np.pi*quad(lambda x: spherical_jn(0,k*x)*x**2.*(np.exp(-x/R_max)-1./np.e),0,R_max)[0]#(1./(1+x)**2.-1./(1+R_max)**2.),0,R_max)[0]
ft_window_hat = lambda k: 4.*np.pi*quad(lambda x: spherical_jn(0,k*x)*x**2.*(1.),0,R_max)[0]
```

```python
k_all = np.logspace(-3,0,100)
fts=[ft_window(kk)/ft_window(0) for kk in k_all]
fts_hat=[ft_window_hat(kk)/ft_window_hat(0) for kk in k_all]
plt.plot(k_all,fts_hat)
plt.plot(k_all,fts)
```

#### Scratch hankel transform code

```python
def hankel_transformand(x,k,R_max):
    return r2xi_spline(x)*window(x,R_max)/(k*x)**(1.5)/x
def hankel_transformand_infinity(x,k):
    return r2xi_spline(x)/(k*x)**1.5/x
def integrated_kernel(k_min,k_max,R_max):
    ht=HankelTransform(nu=3/2,h=k_min/100)
    min_int=ht.transform(lambda x: hankel_transformand(x,k_min,R_max))[0]
    max_int=ht.transform(lambda x: hankel_transformand(x,k_max,R_max))[0]
    return (max_int-min_int)*np.sqrt(np.pi/2.)*3./(k_max**3.-k_min**3.)
def integrated_kernel_infinity(k_min,k_max):
    ht=HankelTransform(nu=3/2,h=k_min/100)
    min_int=ht.transform(lambda x: hankel_transformand_infinity(x,k_min))[0]
    max_int=ht.transform(lambda x: hankel_transformand_infinity(x,k_max))[0]
    return (max_int-min_int)*np.sqrt(np.pi/2.)*3./(k_max**3.-k_min**3.)
def integrated_kernel_thin(k_cen,R_max):
    ht=HankelTransform(nu=1/2,h=k_cen/100)
    return np.sqrt(np.pi/(2.*k_cen))*ht.transform(lambda x: r2xi_spline(x)*window(x,R_max)/x**(1.5),k_cen)[0]
def integrated_kernel_thin_infinity(k_cen):
    ht=HankelTransform(nu=1/2,h=k_cen/100)
    return np.sqrt(np.pi/(2.*k_cen))*ht.transform(lambda x: r2xi_spline(x)/x**(1.5),k_cen)[0]
```
